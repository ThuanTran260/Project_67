## Bước 2 — Tạo cấu trúc thư mục

```python
import os, shutil

BASE = '/content/nhom67'
DRIVE = '/content/drive/MyDrive/nhom67_tuan2'

dirs = [
    f'{BASE}/src',
    f'{BASE}/data/raw',
    f'{BASE}/data/processed',
    f'{BASE}/results',
]
for d in dirs:
    os.makedirs(d, exist_ok=True)
    print(f'  ✓ {d}')

print('\n✓ Tạo thư mục xong')
```

## Bước 3 — Chép file từ Drive vào Colab

```python
# Danh sách file cần chép: (nguồn Drive, đích Colab)
files_to_copy = [
    ('runner.py',            f'{BASE}/src/runner.py'),
    ('mbpp_subset.json',     f'{BASE}/data/raw/mbpp_subset.json'),
]

for fname, dest in files_to_copy:
    src = f'{DRIVE}/{fname}'
    if os.path.exists(src):
        shutil.copy(src, dest)
        print(f'  ✓ {fname} → {dest}')
    else:
        print(f'  ✗ KHÔNG TÌM THẤY: {src}')
        print(f'    → Kiểm tra lại tên thư mục trên Drive')

print('\n✓ Chép file xong')
```
